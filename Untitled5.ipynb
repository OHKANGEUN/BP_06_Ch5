2.어떤 사람이 복리이자율 7%로 1000만원을 저금했을 경우에 2000만원이 되는데 몇년이 걸리는지 계산하기 위해 다음과 같은 코드를 작성했다.잘못된점 없는지 체크해보자

```python
year=0
balance=1000


while balance <=2000 :
  year=year+1
  interest=balance*0.07
  balance+balance+interest
  print(year,"년이 걸립니다.")
```

3.다음 코드의 출력을 예상해봐.각단계에 변수 값을 예상해보시오.

```python
n=1234
sum=0
while n>0:
  digit=n%10
  sum=sum+digit
  n=n//10
print(sum)
```

5.사용자가 입력한 정수의 합을 계산하는 프로그램을 작성하자.사용자가 0을 입력하기전까지 정수를 계쏙하여 읽도록 한다.

```python
sum=0
while True:
  x=int(input("정수를 입력하시오: "))
  if x ==0:
    break;
    sum=sum+x

print("합은 ", sum, "입니다.")
```
